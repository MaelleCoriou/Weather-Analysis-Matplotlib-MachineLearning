# Weather Matplotlib Scatter Plots

```
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime as dt

link = "https://raw.githubusercontent.com/murpi/wilddata/master/quests/weather2019.csv"
df_weather = pd.read_csv(link).iloc[:,[0, 1, 2, -3]]

df_weather['DATE'] = pd.to_datetime(df_weather['DATE'])
```

# Now, complete the following tasks :

## ScatterPlot with 2 variables, title and legend
First of all, you have to create a scatterplot with MAX (in red) and MIN (in blue) Temperatures on the same graph

X-axis stay "DATE" for both values

You have to add title and legend

Here is an example of what is expected :

matplotlib1.png

```

ax1 = plt.scatter(x = df_weather["DATE"], y = df_weather["MAX_TEMPERATURE_C"], c= 'lightcoral', label="Max")
ax2 = plt.scatter(df_weather["DATE"], y = df_weather["MIN_TEMPERATURE_C"], c= 'paleturquoise', label="Min")
# Légende
plt.legend(title="Températures", handles=[ax1, ax2], loc='upper left')
# Titre graphique
plt.title("Températures Min et Max")
# Titre ordonnée
plt.ylabel("Températures")

plt.show()
```

## 2 charts on the same figure
Now, complete this code to have MAX and MIN Temperature in 2 charts, MAX on the left side and MIN on the right side 

The X axis have to be the date of the year

Each graph must have a title, a legend for the colors, and of course, readable axis (ticks)

Here is an example of what is expected :


Capture d’écran 2021-08-26 à 17.27.31.png

```
fig, (ax1,ax2) = plt.subplots(nrows=1,ncols=2,figsize=(17,5))

# Définition du premier graphique
ax1 = plt.subplot(121) 
ax1.scatter(x = df_weather["DATE"], y = df_weather["MAX_TEMPERATURE_C"], c= 'lightcoral', label="Max")
# Titre graphique
ax1.set_title("Température Max")
# Titre ordonnée
ax1.set_ylabel("Température Max")
# Echelle ordonnée
ax1.set_ylim([-5, 40])

# Définition du second graphique
ax2 = plt.subplot(122) 
ax2.scatter(df_weather["DATE"], y = df_weather["MIN_TEMPERATURE_C"], c= 'steelblue', label="Min")
# Titre graphique
ax2.set_title("Température Min")
# Titre ordonnée
ax2.set_ylabel("Température Min")
# Echelle ordonnée
ax2.set_ylim([-5, 40])

# Afficher les graphs
plt.show()
```
